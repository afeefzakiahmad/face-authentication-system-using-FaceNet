## Drag and drop images in the folder - Images/NewFaces, and run this script to generate embeddings

```python
from src.model import InceptionResNetV1
from src.face_auth import getEmbedding
import os
import numpy as np


import sys
log_file = open('log.txt', 'r')
sys.stderr = log_file
```

```python
weights_path = 'src/facenet_keras_weights.h5'

model = InceptionResNetV1(weights_path = weights_path)

print("Model loaded successfully.")
```

```python
embed_stored_path = 'src/embeddings.npz'

if os.path.isfile(embed_stored_path):
    data = np.load(embed_stored_path)
    embeddings = {id:data[id] for id in data.files}
else:
    embeddings ={}
```

```python
img_folder = 'Images\\NewFaces'
filename = os.listdir(img_folder)
for i in range(len(filename)):
    if filename[i] in embeddings:
        print("Person is already enrolled.")
        print("Override Embedding?(Y/N) ")
        ans = input()
        if ans.lower() == 'n':
            print("Quiting...")
            break
        elif ans.lower() == 'y':
            image_path = os.path.join(img_folder,filename[i])
            embed = getEmbedding(model,image_path)
            embeddings.update({filename[i]:embed})
        else:
            print("Wrong Input. Quiting")
            break
    
    else:
        image_path = os.path.join(img_folder,filename[i])
        embed = getEmbedding(model,image_path)
        embeddings.update({filename[i]:embed})
```
