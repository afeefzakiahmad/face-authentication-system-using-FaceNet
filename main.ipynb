# Use this section for Face Recognition (1 to N)
## Enter path of the image to recognize

```python
img_path = 'Images/val/n001153/3.jpg'
assert os.path.isfile(img_path), "Image path is incorrect."
```

```python
from src.face_auth import getEmbedding, l2_loss
import numpy as np



newface = getEmbedding(model, img_path)

data = np.load('src/embeddings.npz')
embeddings = {id:data[id] for id in data.files}

def recognize(newface, embeddings):
    for id in embeddings:
        loss = l2_loss(newface, embeddings[id])
        if loss < 0.8:
            return f"Welcome! Your ID is {id[:-3]} Loss is {loss}"
    return f"Not Authorized! Loss is {loss}"


print(recognize(newface, embeddings))
```
